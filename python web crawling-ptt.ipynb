```python
import requests
from bs4 import BeautifulSoup
article_url = "Insert page url here"
response = requests.get(article_url)
soup = BeautifulSoup(response.text, "lxml")

#獲取一篇文章的內文
def get_article_title(url):
    title=[]
    content = soup.select("span.article-meta-value")
    for i in content:
        title.append(i.text)
    return title
    
        
def checkformat(soup, class_tag, data, index, url):
    content = soup.select(class_tag)[index].text
    return content



def get_article_content(url):
    content = soup.find(id="main-content").text
    target_content = u'※ 發信站: 批踢踢實業坊(ptt.cc),'
    date = checkformat(soup, '.article-meta-value', 'date', 3, url)
    
    #擷取target_content以上及date以下所夾的內文
    content = content.split(target_content)
    content = content[0].split(date)
    #去除掉文末 --
    main_content = content[1].replace('--', '')
    return main_content

#獲取一篇文章的留言
def get_article_comments(url):
    comments = []
    content = soup.select("span.f3.push-content")
    for i in content:
        comments.append(i.text.replace(":"," ").strip())
    return comments
        
    

#寫入檔案
with open('movie_message.txt','w') as f:
    extracted_title = get_article_title(article_url)
    for i in extracted_title:
        f.write(i + "\n")
    extracted_content = get_article_content(article_url)
    f.write(extracted_content+ "\n")
    extracted_comments = get_article_comments(article_url)
    for i in extracted_comments:
        f.write(i + "\n")

#獲取一個ptt索引頁面中的所有文章網址
def get_all_title(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml") #用lxml加速解析
    results = soup.select("div.title")
    for item in results:
        a_item = item.select_one("a")
        if a_item:
            get_article_content(article_url='https://www.ptt.cc'+a_item.get('href'))
    print('------------------ next page ------------------')
```

```python
#範例 
article_url = "https://www.ptt.cc/bbs/PlayStation/M.1583584492.A.065.html"
get_article_title(article_url)
get_article_content(article_url)
get_article_comments(article_url)
```
