```python
import requests
from bs4 import BeautifulSoup
article_url = "https://www.ptt.cc/bbs/graduate/M.1583550777.A.224.html"
response = requests.get(article_url)
soup = BeautifulSoup(response.text, "lxml")

#獲取一篇文章的內文
def get_article_title(url):
    content = soup.select("span.article-meta-value")
    for i in content:
        print(i.text)
        
def checkformat(soup, class_tag, data, index, url):
    content = soup.select(class_tag)[index].text
    return content



def get_article_content(url):
    content = soup.find(id="main-content").text
    target_content = u'※ 發信站: 批踢踢實業坊(ptt.cc),'
    date = checkformat(soup, '.article-meta-value', 'date', 3, "https://www.ptt.cc/bbs/graduate/M.1583550777.A.224.html")
    
    #擷取target_content以上及date以下所夾的內文
    content = content.split(target_content)
    content = content[0].split(date)
    #去除掉文末 --
    main_content = content[1].replace('--', '')
    #印出內文
    print(main_content)

#獲取一篇文章的留言
def get_article_comments(url):
    content = soup.select("span.f3.push-content")
    for i in content:
        print(i.text.replace(":"," ").strip())
    

get_article_title(article_url)
get_article_content(article_url)
get_article_comments(article_url)


#獲取一個ptt索引頁面中的所有文章網址
def get_all_title(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml") #用lxml加速解析
    results = soup.select("div.title")
    for item in results:
        a_item = item.select_one("a")
        if a_item:
            get_article_content(article_url='https://www.ptt.cc'+a_item.get('href'))
    print('------------------ next page ------------------')
```
